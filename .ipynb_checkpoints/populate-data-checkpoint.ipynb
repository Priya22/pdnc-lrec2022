```python
def get_arr(s):
    if isinstance(s, str):
        return eval(s)
    elif isinstance(s, list):
        return s
    else:
        raise TypeError
```

```python
def verify_spans(df, ntext):
    qcount = 0
    mcount = 0
    
    qtot = 0
    mtot = 0
    
    for _, row in df.iterrows():
        qtexts = get_arr(row['qTextArr'])
        qspans = get_arr(row['qSpan'])

        mtexts = get_arr(row['menTexts'])
        mspans = get_arr(row['menSpans'])

        for qt, qs, mts, mss in zip(qtexts, qspans, mtexts, mspans):
            qtot += 1
            et = ntext[qs[0]:qs[1]]
            assert et.strip() == qt.strip(), print("mismatch at span {} and text {}".format(qs, qt))
            qcount += 1
            
            for mt, ms in zip(mts, mss):
                mtot += 1
                emt = ntext[ms[0]:ms[1]]
                assert mt.strip() == emt.strip(), print("mismatch at span {} and text {}".format(ms, mt))
                mcount += 1
    
    print("Verified {}/{} quotes".format(qcount, qtot))
    print("Verified {}/{} mentions".format(mcount, mtot))
    
```

```python
verify_spans(qdf, ntext)
```

```python
novels = []
for folder in os.scandir('/h/vkpriya/data/pdnc/'):
    if os.path.isdir(folder) and folder.name[0] not in ['.', '_']:
        novels.append(folder.name)
novels = sorted(novels)
```

```python
print(novels), len(novels)
```

```python
for novel in novels:
    print(novel)
    
    os.makedirs(os.path.join('data/', novel), exist_ok=True)
    
    qtdf = pd.read_csv('/h/vkpriya/data/pdnc/'+novel+'/data_df.csv', index_col=0)
    charDict = pkl.load(open('/h/vkpriya/data/pdnc/'+novel+'/charDict.pkl', 'rb'))
    with open('/h/vkpriya/data/pdnc/'+novel+'/full_text.txt', 'r') as f:
        ntext = f.read().strip()
    
    qdf = qtdf[keep_columns]
    
    verify_spans(qdf, ntext)
    
    qdf.to_csv(os.path.join('data/', novel, 'quotations.csv'), index=False)
    with open(os.path.join('data/', novel, 'text.txt'), 'w') as f:
        print(ntext, file=f)
        
    cDict = {'id2names': charDict['main_id2names'], 'name2id': charDict['main_name2id'], 'id2parent': charDict['main_id2parent']}
    
    pkl.dump(cDict, open(os.path.join('data/', novel, 'charDict.pkl'), 'wb'))
```
